```python
import matplotlib.pyplot as plt

# Create a histogram for the "year" column
plt.figure(figsize=(10, 6))
plt.hist(DFclean['year'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Distribution of Vehicle Manufacturing Years')
plt.grid(True)
plt.show()
```

In this step, we employed the Interquartile Range (IQR) method to identify outliers in the "year" column. The IQR method is a robust statistical technique commonly used for outlier detection.


```python
import numpy as np

def iqr_outliers(column):
    # Calculate the first quartile (Q1) and third quartile (Q3)
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = (column < lower_bound) | (column > upper_bound)
    
    return outliers

# Detect outliers using the IQR method for 'year'
year_outliers_iqr = iqr_outliers(DFclean['year'])

# Count the number of outliers in the 'year' column
year_outlier_count_iqr = year_outliers_iqr.sum()

print(f'Number of outliers in "year" using IQR method: {year_outlier_count_iqr}')
```

Following the identification of outliers in the "year" column using the Interquartile Range (IQR) method, we have taken a proactive approach to address these outliers. Outliers can potentially skew our analysis and distort results, making it important to handle them appropriately.

In this step, we have chosen to convert the identified outliers to NaN (Not-a-Number) values. This decision has been made for several reasons:

- **Data Integrity**: Converting outliers to NaN values helps maintain the integrity of the dataset. Outliers are often erroneous or inconsistent data points that can adversely impact the quality of our analysis.

- **Consistency**: By replacing outliers with NaNs, we ensure that the data remains consistent. These values are likely to be inconsistent with the majority of the dataset.

- **Transparency**: This approach is transparent and clearly documents the handling of outliers. It provides a clear record of the decisions made during data preprocessing.

Converting outliers to NaNs allows us to handle them more effectively in subsequent steps of our analysis, such as filtering, imputation, or exclusion, depending on the specific requirements of our analysis.

```python
# Convert identified outliers in the "year" column to NaN
DFclean.loc[year_outliers_iqr, 'year'] = np.nan
```

```python
def iqr_outliers(column):
    # Calculate the first quartile (Q1) and third quartile (Q3)
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = (column < lower_bound) | (column > upper_bound)
    
    return outliers

# Detect outliers using the IQR method for 'odometer'
odometer_outliers_iqr = iqr_outliers(DFclean['odometer'])

# Count the number of outliers in the 'odometer' column
odometer_outlier_count_iqr = odometer_outliers_iqr.sum()

print(f'Number of outliers in "odometer" using IQR method: {odometer_outlier_count_iqr}')
```

```python
# Convert identified outliers in the "odometer" column to NaN using .loc
DFclean.loc[odometer_outliers_iqr, 'odometer'] = np.nan
```

2. **Acquisition Prices**: The initial purchase price of each vehicle is a significant factor. Cars with higher acquisition costs should ideally generate more profit over their lifetime. However, it's essential to assess whether these high-cost vehicles are delivering the expected returns.

3. **Profit Margins**: Calculate the profit margins associated with each vehicle. This involves subtracting the total expenses (including maintenance and acquisition costs) from the revenue generated by leasing or selling the car. Focus on cars that provide higher profit margins and consider retaining or investing in similar models.

4. **Resale Value**: Evaluate the expected resale value of each vehicle. Some cars may retain their value better than others, making them more appealing for resale or leasing.

5. **Market Demand**: Assess the current market demand for different car models. Vehicles that are in high demand are more likely to yield better prices. Conversely, cars with low demand may need to be sold at a discount.

7. **Customer Feedback**: Solicit feedback from customers who have leased or used your cars. Their experiences and preferences can inform decisions about which models to retain or invest in.

In conclusion, making informed decisions about your car fleet's composition requires a comprehensive analysis that considers both financial and operational aspects. ### Predicting Prices for Cars with Missing Values

In our data analysis journey, we encountered a subset of cars with a price value of 0. While we've taken extensive steps to clean and enhance our dataset, there are instances where some cars still lack accurate pricing information. In such cases, we face the challenge of determining a reasonable estimate for these missing prices.

#### Why Are Some Prices Missing?

The absence of pricing information in a used car dataset can stem from various reasons. Some cars may genuinely be priced at zero, indicating that they are not intended for sale, while others may have errors or omissions in their listing details. Regardless of the cause, it's crucial to address these missing values to ensure the dataset's integrity and usefulness.

#### A Last Resort: Predictive Modeling

As a last resort, we can employ predictive modeling techniques to estimate prices for the cars with missing values. This involves using the available data, including features like year, mileage (odometer), and condition, to create a model that can make price predictions. While this approach can provide approximate values, it's important to note that these predictions may not be highly accurate, especially for cars with unique or unusual characteristics.

#### The Importance of Manual Review

Before resorting to predictive modeling, we strongly recommend a comprehensive manual review of the cars with missing prices. This involves investigating each listing individually to determine if there are data entry errors, incomplete listings, or specific circumstances that explain the missing price. Manual review can often provide more context and accuracy than automated modeling.

#### Transparency and Cautious Decision-Making

In situations where pricing information is unclear, it's essential to be transparent about the limitations of any predictive models used. While these models can provide guidance, they should not replace a diligent review of each car's unique circumstances. Additionally, any decisions made based on predicted prices should be approached with caution, especially when dealing with a substantial number of cars with missing values.

In summary, predicting prices for cars with missing values is a useful tool when no other information is available. However, manual review and data verification should always be the preferred approach to ensure accuracy and reliability in decision-making.

```python
# DataFrame con coches que tienen precio diferente de 0
df_con_precio = DFclean_no_outliers_nans[DFclean_no_outliers_nans['price'] != 0]

# DataFrame con coches que tienen precio igual a 0
df_sin_precio = DFclean_no_outliers_nans[DFclean_no_outliers_nans['price'] == 0]
```

```python
from sklearn.linear_model import LinearRegression
# Dividir los datos en caracterÃ­sticas (X) y etiquetas (y)
X = df_con_precio[['year', 'odometer']]  # Las caracterÃ­sticas que utilizaremos para la predicciÃ³n
y = df_con_precio['price']  # El valor de precio conocido

# Crear un modelo de regresiÃ³n lineal
modelo_regresion = LinearRegression()

# Entrenar el modelo con los datos conocidos
modelo_regresion.fit(X, y)

# Utilizar el modelo para predecir los precios de los coches con precio igual a 0
X_sin_precio = df_sin_precio[['year', 'odometer']]
predicciones_precio = modelo_regresion.predict(X_sin_precio)

# Asignar las predicciones de precio de vuelta al DataFrame original usando .loc[]
df_sin_precio.loc[:, 'price'] = predicciones_precio


# Combinar los DataFrames nuevamente si es necesario
df_final = pd.concat([df_con_precio, df_sin_precio])

# Ahora, df_final contiene los valores de precio predichos para los coches que tenÃ­an precio igual a 0
```
